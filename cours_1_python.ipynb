### Listes (type list)

```python
liste_vide = []
liste_vide_v2 = list()
groupe_de_5 = [n1, n2, n3, n4]
```

```python
print(liste_vide == liste_vide_v2)
```

```python
print(groupe_de_5)
```

```python
# L'indice du premier élement de la liste est 0
# L'indice du premier élement de la liste est -1 ou N - 1, N étant la longueur de la liste
print("Le premier élement de la liste est", groupe_de_5[0])
print("Le dernier élement de la liste est", groupe_de_5[-1])
print("Le deuxième élement de la liste est", groupe_de_5[1])
print("Le troisième élement de la liste est", groupe_de_5[2])
```

```python
# append permet d'ajouter un élement à la fin de la liste
groupe_de_5.append("cinq")
```

```python
print(groupe_de_5)
```

```python
# remove permet d'enlever la première occurence d'un élement dans la liste
groupe_de_5.remove(5)
```

```python
print(groupe_de_5)
```

```python
# pop permet d'enlever et d'afficher le ième élément de la liste
groupe_de_5.pop(2)
```
